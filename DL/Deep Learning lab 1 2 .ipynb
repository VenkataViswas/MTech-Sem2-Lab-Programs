# 1.1	Design a simple perceptron with two inputs and random weights

```python
import random
def perceptron(x,wt):
  f = 0
  for i,w in zip(x,wt):
    f+=(i*w)
  if f>=1 :
    return 1
  else: return  0
```

```python
x = [5,0.2]
wt = random.sample(range(-1,2),2)
print("input : ", x, "\n weights random : ",wt)
print(" output : ", perceptron(x,wt))
```

#1.2 Perceptron with multiple inputs and random weights

```python
import random
def perceptron(x,wt):
  f = 0
  for i,w in zip(x,wt):
    w = w/10
    f+=(i*w)
  if f>=1 :
    return 1
  else: return  0
```

```python
x = [ 7.8,2,2,0.5]
wt = random.sample(range(-1,len(x)),len(x))
print("input : ", x, "\n weights random : ",wt)
print(" output : ", perceptron(x,wt))
```

1.4 Perceptron with 2 inputs and 1 output

```python
def predict(row, weights):
    total = weights[0]  # bias
    for i in range(len(row) - 1):
        total += weights[i + 1] * row[i]
    return 1 if total >= 0 else 0


dataset = [
    [2, 2, 0],
    [1, 2, 0],
    [5, 4, 1],
    [2, 1, 0],
    [3, 3, 1],
    [7, 2, 1],
    [5, 2, 1],
    [6, 1, 1],
    [8, -0.2, 1],
    [7, 3, 1]
]

weights = [-0.4, 0.99999999992, 0.10000000003]

for row in dataset:
    result = predict(row, weights)
    print("Expected =", row[-1], ", Predicted =", result)
```

# 1.5 1.5	Implement a perceptron to classify a movie by training the weights with stochastic gradient descent model

```python
# Make a prediction with wt; Estimate weights with Stochastic gradient descent
def predict(row, wt):
    activation = wt[0]
    for i in range(len(row) - 1):
        activation += wt[i + 1] * row[i]
    return 1.0 if activation >= 0.0 else 0.0
```

```python


# Estimate Perceptron wt using stochastic gradient descent
def train_wt(train, l_rate, n_turn):
    wt = [0.0 for i in range(len(train[0]))]
    for turn in range(n_turn):
        sum_error = 0.0
        for row in train:
            prediction = predict(row, wt)
            error = row[-1] - prediction
            sum_error += error**2
            wt[0] = wt[0] + l_rate * error
            for i in range(len(row) - 1):
                wt[i + 1] = wt[i + 1] + l_rate * error * row[i]
        print('>turn=%d, lrate=%.3f, error=%.3f' % (turn, l_rate, sum_error))
    return wt
```

```python


# Calculate wt
dataset = [
    [2, 2, 0],
    [1, 2, 0],
    [5, 4, 1],
    [2, 1, 0],
    [3, 3, 1],
    [7, 2, 1],
    [5, 2, 1],
    [6, 1, 1],
    [8, -2, 1],
    [7, 3, 1]
]

l_rate = 0.1
n_turn = 4
wt = train_wt(dataset, l_rate, n_turn)
print(wt)

for row in dataset:
    prediction = predict(row, wt)
    print("Expected=%d, Predicted=%d" % (row[-1], prediction))

```

2  Feed Forward network with multiple neurons

```python
# Sigmoid activation function
import math

def sigmoid(x):
    return 1.0 / (1.0 + math.exp(-x))


# Compute output of a single neuron
def neuron_output(weights, inputs):
    total = weights[0]  # bias
    for i in range(len(inputs)):
        total += weights[i + 1] * inputs[i]
    return sigmoid(total)


# Forward propagation
def forward_propagate(ffnet_layers, row):
    inputs = row[:-1]  # exclude label
    print("Input:", inputs)

    for layer in ffnet_layers:
        new_inputs = []
        for neuron in layer:
            output = neuron_output(neuron, inputs)
            new_inputs.append(output)
        inputs = new_inputs

    return inputs[0]  # final output


# Feed Forward Network definition
ffnet_layers = [
    # Input → Hidden layer (2 neurons)
    [
        [0.1, 0.2, -0.3],   # neuron 1: bias, w1, w2
        [-0.1, 0.4, 0.1]   # neuron 2: bias, w1, w2
    ],
    # Hidden → Output layer (1 neuron)
    [
        [0.05, 0.3, -0.2]  # bias, w_hidden1, w_hidden2
    ]
]


# Dataset
dataset = [
    [1, 2, 0],
    [5, 4, 1],
    [2, 1, 0],
    [3, 3, 1],
    [7, 2, 1]
]


# Run forward propagation
for row in dataset:
    output = forward_propagate(ffnet_layers, row)
    prediction = 1 if output >= 0.5 else 0
    print(f"Expected={row[-1]}, Predicted={prediction}, Output={output:.3f}")
```
