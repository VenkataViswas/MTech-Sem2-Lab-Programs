#Create Fuzzy relation by Cartesian product of any two Fuzzy sets

```python
import pandas as pd

def fuzzy_relation(A, B):

    rows = list(A.keys())
    columns = list(B.keys())

    # Initialize DataFrame
    df = pd.DataFrame(index=rows, columns=columns)

    # Fill the matrix using min operation
    for a in rows:
        for b in columns:
            df.loc[a, b] = min(A[a], B[b])

    return df
```

```python
A = {'a1': 0.2, 'a2': 0.7, 'a3': 0.4}
B = {'b1': 0.5, 'b2': 0.6}

relation = fuzzy_relation(A, B)
display(relation)
```

## Perform Max-Min composition of any two Fuzzy relations(with different conditions



```python
import pandas as pd

def max_min_composition(R, S):
    X = R.index
    Y = R.columns
    Z = S.columns

    # Initialize result relation
    T = pd.DataFrame(0.0, index=X, columns=Z)

    # Max–Min composition
    for x in X:
        for z in Z:
            T.loc[x, z] = max(
                min(R.loc[x, y], S.loc[y, z]) for y in Y
            )

    return T
```

```python
# R: X → Y
R = pd.DataFrame(
    [[0.2, 0.7],
     [0.5, 0.4]],
    index=['x1', 'x2'],
    columns=['y1', 'y2']
)

# S: Y → Z
S = pd.DataFrame(
    [[0.6, 0.3],
     [0.8, 0.5]],
    index=['y1', 'y2'],
    columns=['z1', 'z2']
)

T = max_min_composition(R, S)
display(T)
```
