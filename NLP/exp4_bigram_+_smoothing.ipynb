Bi-Gram Model and Add-one Smoothing

```python
def readData():
  data = ["this is a dog","this is a cat", "I love my cat","this is my name"]
  d =[]
  for sent in data :
    for word in sent.split() :
      d.append(word)
  print(d)
  return d
```

```python
def createBigram(data):
    listOfBigrams = []
    bigramCounts = {}
    unigramCounts = {}

    for i in range(len(data) - 1):
        if data[i + 1].islower():
            bigram = (data[i], data[i + 1])
            listOfBigrams.append(bigram)

            bigramCounts[bigram] = bigramCounts.get(bigram, 0) + 1
            unigramCounts[data[i]] = unigramCounts.get(data[i], 0) + 1

    return listOfBigrams, unigramCounts, bigramCounts
```

```python


def calcBigramProb(unigramCounts, bigramCounts):
    bigramProb = {}
    V = len(unigramCounts)   # vocabulary size (for smoothing)

    for (w1, w2), count in bigramCounts.items():
        bigramProb[(w1, w2)] = (count + 1) / (unigramCounts[w1] + V)

    return bigramProb

```

```python

# ---------------- MAIN ----------------
if __name__ == '__main__':

    data = readData()   # training data (string)

    listOfBigrams, unigramCounts, bigramCounts = createBigram(data)

    print("\nAll the possible Bigrams:")
    print(listOfBigrams)

    print("\nBigrams along with their frequency:")
    print(bigramCounts)

    print("\nUnigrams along with their frequency:")
    print(unigramCounts)

    bigramProb = calcBigramProb(unigramCounts, bigramCounts)

    print("\nBigrams along with their probability (smoothed):")
    print(bigramProb)

    # Sentence probability
    inputSentence = "This is my cat"
    splt = inputSentence.split()

    bilist = []
    for i in range(len(splt) - 1):
        bilist.append((splt[i], splt[i + 1]))

    print("\nThe bigrams in given sentence are:")
    print(bilist)

    V = len(unigramCounts)
    outputProb1 = 1

    for (w1, w2) in bilist:
        if (w1, w2) in bigramCounts:
            outputProb1 *= (bigramCounts[(w1, w2)] + 1) / (unigramCounts[w1] + V)
        else:
            outputProb1 *= 1 / (unigramCounts.get(w1, 0) + V)

    print('\nProbability of sentence "This is my cat" =', outputProb1)
```
